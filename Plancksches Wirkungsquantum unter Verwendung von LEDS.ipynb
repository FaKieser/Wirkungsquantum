```python
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 1. Bestimmung der Schwellspannung für die LEDs

## 1.1 Berechnung der Auslgleichsgerade für die rote LED

```python
Strom_rot = [0.08,0.55,0.77,5.6,16.7,26.39,57.1]
Spannung_rot = [0.97,1.13,1.17,1.51,1.72,1.82,2]

plt.plot(Spannung_rot, Strom_rot, 'ro', label='Messwerte')
plt.xlabel('Spannung [V]')
plt.ylabel('Strom [mA]')
plt.grid()
plt.show
```

```python
model = LinearRegression()
x_r = np.array(Spannung_rot[4:7])
y_r = np.array(Strom_rot[4:7])

x_r = x_r.reshape(-1,1)

model.fit(x_r,y_r)
b_r = model.intercept_
m_r = model.coef_

print("y-Achsen-Abschnitt:", model.intercept_)
print("Steigung:", model.coef_)
print("R²:", model.score(x_r, y_r))
```

```python
# Einen x-y-Plot erstellen:
Strom_rot = [0.08,0.55,0.77,5.6,16.7,26.39,57.1]
Spannung_rot = [0.97,1.13,1.17,1.51,1.72,1.82,2]

X_r = np.arange(1.4, 2.2, 0.05)
plt.plot(Spannung_rot, Strom_rot, 'ro', label='Messwerte')
plt.plot(X_r, m_r*X_r+b_r, '--k', label = 'Ausgleichsgerade')
plt.legend()
plt.xlabel('Spannung [V]')
plt.ylabel('Strom [mA]')
plt.grid()
plt.show
```

# 2. Berechnung des Planckschen Wirkungsquantums

```python
def h(U_s, lampda):
    c = 3*(10**8)
    e = 1.6*(10**(-19))
    return U_s* e* lampda*(10**-9)/c
```

```python
U_S = [1.6, 1.6, 1.8]
lampda = [630, 520, 590]
```

```python
t = []
for i in range(3):
    t.append(h(U_S[i],lampda[i]))
t
```

```python
p = []
for i in range(3):
    proz = (1 - t[i] / 6.6 * 10**34) *100
    p.append(proz)
p
```

```python
t_1 = []
for i in range(3):
    T = t[i] * 10**34
    t_1.append(T)
t_1
```

# 3. Darstellung der Daten

```python
df = pd.DataFrame([lampda, t_1, p], index = pd.Index(['Wellenlänge $\lambda$ in nm', '$h = \dfrac{e \cdot U_s \cdot \lambda }{c}$ in $10^{-34} \, \mathrm{W \cdot s^2}$', 'Abweichung in Prozent']), 
                  columns = pd.Index(['Rot', 'Gelb', 'Blau'], name = 'Farbe'))
df.round(2)
```
